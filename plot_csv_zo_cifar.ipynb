```python
import numpy as np
import matplotlib.pyplot as plt
import os
print(os.getcwd())
```

```python
def get_error(path,loss):
    error=0
    for n in range(0,20):
        loss_n=0
        for k in range(0,10):
            filename= path+'/class_'+str(k)+str(n)+'_loss.csv'
            file = open(filename)
            loss_n += np.loadtxt(file, delimiter=",")
        error+=(loss_n/10-loss)**2
    return np.sqrt(error/20)
```

```python
file = open("experiment_results/CIFER_PN_solvers.zo.ada_exp_grad_2.0/average_loss.csv")
adaexpmdpn = np.loadtxt(file, delimiter=",")
```

```python
fig, ax=plt.subplots()
ax.plot(range(0,201), adaexpmdpn,label=r'ZO-AdaExpGrad', marker='v',markevery=10)
plt.xlabel("Iteration")
plt.ylabel("loss")
ax.set_yscale("log")
plt.legend()
plt.show()
```
